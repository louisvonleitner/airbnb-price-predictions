```python
import matplotlib.pyplot as plt

# Sum each one-hot column (each amenity)
amenity_counts = X.sum().sort_values(ascending=False)

# Plot top 20
plt.figure(figsize=(12, 6))
amenity_counts.head(20).plot(kind='bar')
plt.title("Top 20 Most Common Amenities")
plt.xlabel("Amenity")
plt.ylabel("Number of Apartments")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
